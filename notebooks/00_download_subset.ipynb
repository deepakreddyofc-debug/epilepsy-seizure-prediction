```python
import os
import requests

base_url = "https://physionet.org/files/chbmit/1.0.0/"
save_dir = "../data_raw/chb_subset"
os.makedirs(save_dir, exist_ok=True)

subjects = ["chb01", "chb02", "chb03"]  # small subset

for subj in subjects:
    print(f"\nDownloading {subj}...")
    subj_dir = os.path.join(save_dir, subj)
    os.makedirs(subj_dir, exist_ok=True)

    index_url = f"{base_url}{subj}/"
    html = requests.get(index_url).text

    edf_files = [line.split('"')[1] for line in html.splitlines() if ".edf" in line]

    for f in edf_files:
        file_url = f"{index_url}{f}"
        save_path = os.path.join(subj_dir, f)

        if os.path.exists(save_path):
            print("Already exists:", f)
            continue

        print("Downloading:", f)
        with requests.get(file_url, stream=True) as r:
            r.raise_for_status()
            with open(save_path, "wb") as fp:
                for chunk in r.iter_content(chunk_size=8192):
                    if chunk:
                        fp.write(chunk)

print("\nSubset download complete!")
```
