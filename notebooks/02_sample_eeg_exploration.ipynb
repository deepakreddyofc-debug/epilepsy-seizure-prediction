```python
import os
import urllib.request

# URL of the CHB-MIT dataset (zip file)
url = "https://physionet.org/static/published-projects/chbmit/chbmit-1.0.0.zip"

# Location to save the downloaded file
save_path = "../data_raw/chbmit.zip"

# Create data_raw folder if not exists
os.makedirs("../data_raw", exist_ok=True)

# Download the dataset
if not os.path.exists(save_path):
    print("Downloading CHB-MIT dataset... (This is ~1.1GB, may take time)")
    urllib.request.urlretrieve(url, save_path)
    print("Download complete! Saved to:", save_path)
else:
    print("Dataset already exists at:", save_path)
```
