```python
import numpy as np
import mne

def load_edf(file_path):
    """Load EDF EEG file using MNE."""
    raw = mne.io.read_raw_edf(file_path, preload=True, verbose=False)
    return raw

def plot_raw(raw, duration=10):
    """Plot EEG for a few seconds."""
    raw.plot(duration=duration, n_channels=19)
```
