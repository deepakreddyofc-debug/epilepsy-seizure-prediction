```python
import numpy as np
from pathlib import Path

processed_dir = Path("../data_processed")
files = sorted(processed_dir.glob("chb01_*_features.npy"))

all_X = []
all_labels = []
all_times = []

window_size_sec = 5
for f in files:
    X_file = np.load(f)
    # labels_file or recompute labels here using seizure intervals + windows
    labels_file = np.load(str(f).replace("_features.npy", "_labels.npy"))
    n_windows = X_file.shape[0]
    times = np.arange(n_windows) * window_size_sec

    all_X.append(X_file)
    all_labels.append(labels_file)
    all_times.append(times)

X = np.vstack(all_X)
labels = np.concatenate(all_labels)
window_times = np.concatenate(all_times)
```
